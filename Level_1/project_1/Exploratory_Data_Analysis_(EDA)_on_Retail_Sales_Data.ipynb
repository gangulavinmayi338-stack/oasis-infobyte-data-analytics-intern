```python
top_calories = df[['Item','Calories']].sort_values(
    by='Calories',
    ascending=False
)

top_calories.head(10)
```

```python
top_protein = df[['Item','Protein']].sort_values(
    by='Protein',
    ascending=False
)

top_protein.head(10)
```

```python
df.groupby('Category')['Calories'].mean()
```

```python
df.groupby('Category')['Protein'].mean()
```

```python
import matplotlib.pyplot as plt

avg_calories = df.groupby('Category')['Calories'].mean()

avg_calories.plot(kind='bar', figsize=(10,5))
plt.title('Average Calories by Category')
plt.ylabel('Calories')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,8))
sns.heatmap(df.select_dtypes(include='number').corr(),
            annot=True,
            cmap='coolwarm')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df['Calories'], df['Protein'])
plt.xlabel('Calories')
plt.ylabel('Protein')
plt.title('Calories vs Protein')
plt.show()
```
