```python
import pandas as pd

df = pd.read_csv('menu.csv')  # replace with actual filename
df.head()
```

```python
print(df.shape)
print(df.columns)
print(df.info())
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
top_calories = df[['Item','Calories']].sort_values(
    by='Calories',
    ascending=False
)

top_calories.head(10)
```

```python
df.groupby('Category')['Calories'].mean()
```

```python
import matplotlib.pyplot as plt

df.groupby('Category')['Calories'].mean().plot(kind='bar')

plt.title('Average Calories by Category')
plt.xlabel('Category')
plt.ylabel('Calories')
plt.show()
```

```python
plt.hist(df['Total Fat'], bins=20)
plt.title('Fat Distribution')
plt.xlabel('Total Fat')
plt.ylabel('Count')
plt.show()
```

```python
plt.scatter(df['Sugars'], df['Calories'])

plt.xlabel('Sugar')
plt.ylabel('Calories')
plt.title('Sugar vs Calories')
plt.show()
```
